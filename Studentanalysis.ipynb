```python
# Student Performance Analysis
# Tools used: Python, Pandas
# Dataset: 20 students across 3 departments
# Goal: Find patterns in student marks and results
```

```python
import pandas as pd

data = {
    "Name": ["Priya", "Sneha", "Ravi", "Anjali", "Meera",
              "Pooja", "Divya", "Kavya", "Amit", "Rahul",
              "Neha", "Sana", "Rohit", "Tina", "Karan",
              "Nisha", "Vivek", "Ritika", "Arjun", "Simran"],
    "Department": ["Science", "Commerce", "Arts", "Science", "Commerce",
                   "Arts", "Science", "Commerce", "Arts", "Science",
                   "Commerce", "Arts", "Science", "Commerce", "Arts",
                   "Science", "Commerce", "Arts", "Science", "Commerce"],
    "Marks": [91, 78, 55, 88, 65, 72, 95, 80, 60, 85,
              70, 58, 92, 75, 63, 89, 77, 68, 94, 82],
    "City": ["Mumbai", "Pune", "Delhi", "Bangalore", "Mumbai",
             "Pune", "Delhi", "Bangalore", "Mumbai", "Pune",
             "Delhi", "Bangalore", "Mumbai", "Pune", "Delhi",
             "Bangalore", "Mumbai", "Pune", "Delhi", "Bangalore"],
    "Result": ["Pass", "Pass", "Fail", "Pass", "Fail",
               "Pass", "Pass", "Pass", "Fail", "Pass",
               "Pass", "Fail", "Pass", "Pass", "Fail",
               "Pass", "Pass", "Pass", "Pass", "Pass"]
}

df = pd.DataFrame(data)
print(df.shape)
```

```python
df.to_csv("student_data.csv", index=False)
print("File saved!")
```

```python
# Step 3: Average marks per department
# Insight: Which department performs best?
dept_avg = df.groupby("Department")["Marks"].mean().round(2)
print(dept_avg)
```

```python
# Step 4: Pass and Fail count
# Insight: How many students passed vs failed?
result_count = df["Result"].value_counts()
print(result_count)
```

```python
# Step 5: Top 5 students by marks
top5 = df.sort_values("Marks", ascending=False).head()
print(top5)
```

```python
# Step 6: Students who failed
# Insight: Which departments have failing students?
failed = df[df["Result"] == "Fail"]
print(failed[["Name", "Department", "Marks"]])
```

```python
# Key Findings:
# 1. Science has highest average marks at 90.57
# 2. Arts has lowest average marks at 62.67
# 3. 5 out of 20 students failed (25%)
# 4. All failing students scored below 70 marks
print("Analysis Complete")
```
